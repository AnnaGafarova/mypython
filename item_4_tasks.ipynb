```python
# постройте фигуру с двумя подграфиками (1 строка, 2 столбца) и визуализируйте:
#   1) Левый подграфик - линейная функция y = x по точкам x из [0, 10], равномерно 50 значений.
#      Добавьте цвет, подпись кривой (label), заголовок, подписи осей, легенду и сетку.
#   2) Правый подграфик - квадратичная функция y = x^2 на тех же x.
#      Задайте иной стиль линии (например, пунктир), цвет и аналогичные подписи/легенду/сетку.
#   3) Добавьте общий заголовок фигуры и аккуратно разнесите элементы макета.
#   4) Сохраните результат в PNG (с dpi=150) и в SVG, затем отобразите график.



#   - Сгенерируйте x через np.linspace(0, 10, 50); вычислите y1 = x, y2 = x**2.
#   - Создайте холст и оси: fig, axes = plt.subplots(1, 2, figsize=(10, 4)).
#   - Левый график: axes[0].plot(x, y1, color=..., label=...); затем set_title, set_xlabel, set_ylabel, legend, grid.
#   - Правый график: axes[1].plot(x, y2, linestyle='--', color=..., label=...); затем те же подписи/легенда/сетка.
#   - Общий заголовок: fig.suptitle(...); компоновка: fig.tight_layout().
#   - Сохранение: fig.savefig("subplots_example.png", dpi=150) и fig.savefig("subplots_example.svg").
#   - Показ: plt.show().
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 50)
y1 = x
y2 = x**2

# 2 подграфика (1 строкаб 2 столбца)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Левый подграфик (y=x)
axes[0].plot(x, y1, color="blue", label="y=x")
axes[0].set_title("Линейная функция")
axes[0].set_xlabel("Ось Х")
axes[0].set_ylabel("Ось Y")
axes[0].legend()
axes[0].grid(True)

# Правый подграфик (y = x**2)
axes[1].plot(x, y2, color="red", linestyle="--", label="y=x**2")
axes[1].set_title("Квадратичная функция")
axes[1].set_xlabel("Ось Х")
axes[1].set_ylabel("Ось Y")
axes[1].legend()
axes[1].grid(True)

# Общий заголовок результульрующего графика (Output data)
fig.suptitle("2 подграфика на одном графике", fontsize=14)
fig.tight_layout()

# Сохраним в PNG и SVG
fig.savefig("subplots_example.png", dpi=150)
fig.savefig("subplots_example.svg")

plt.show()
print("Файлы сохранены subplots_example.png и subplots_example.svg")
```
