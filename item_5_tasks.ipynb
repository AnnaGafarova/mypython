```python
# Постройте простой линейный график функции y = 3x - 2 для x ∈ {-5, -4, ..., 5}.
# Добавьте заголовок, подписи осей, легенду и сетку. Отобразите график на экране.
import numpy as np
import matplotlib.pyplot  as plt

x = np.arange(-5,5)
y = 3*x - 2
plt.plot(x, y, label="y = 3x - 2")

plt.title("Линейный график")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.legend()
plt.grid(True)

plt.show()


# Используйте: range(...) или np.arange(...); plt.plot(..., label=...);
# plt.title(...); plt.xlabel(...); plt.ylabel(...); plt.legend(); plt.grid(True); plt.show().
```

```python
# Создайте фигуру с двумя областями построения (1×2). На первом Axes изобразите y = sin(x),
# на втором - y = cos(x) для x на отрезке [0, 2π] (равномерная сетка точек).
# Подпишите оси, добавьте легенды и сетку. На первом Axes аннотируйте максимум sin(x) стрелкой.
# Добавьте общий заголовок фигуры и выровняйте макет.
import math
x = np.linspace(0, 2*math.pi)


y1 = np.sin(x)
y_max = max(y1)
ind_y_max = np.where(y1 == y_max)
x_max = x[ind_y_max]
y2 = np.cos(x)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Левый подграфик 
axes[0].plot(x, y1, color="blue", label="y=sin(x)")
axes[0].set_title("Функция sin")
axes[0].set_xlabel("Ось Х")
axes[0].set_ylabel("Ось Y")
axes[0].legend()
axes[0].grid(True)
# axes[0].annotate("Максимум sin(x)", xy=(2, 4), xytext=(3, 5),
# arrowprops=dict(arrowstyle="->"))
axes[0].annotate("Максимум sin(x)",xy=(x_max, y_max), xytext=(x_max+1, y_max+1) , arrowprops=dict(arrowstyle="->"))
# Правый подграфик 
axes[1].plot(x, y2, color="red", linestyle="--", label="y=cos(x)")
axes[1].set_title("Функция cos")
axes[1].set_xlabel("Ось Х")
axes[1].set_ylabel("Ось Y")
axes[1].legend()
axes[1].grid(True)

fig.suptitle("2 подграфика на одном графике", fontsize=14)
fig.tight_layout()


plt.show()
# Используйте: np.linspace(...); fig, axes = plt.subplots(1, 2, figsize=...);
# axes[i].plot(..., label=...); axes[i].set_title(...); set_xlabel(...); set_ylabel(...);
# axes[i].grid(True); axes[i].legend(); axes[0].annotate(..., arrowprops=dict(arrowstyle="->"));
# fig.suptitle(...); fig.tight_layout(); plt.show().

```

```python
# На одном графике постройте три функции на [0, 6]: y1 = 2x + 1, y2 = x**1.5, y3 = 0.5x**2.
# Для каждой линии задайте свой цвет, стиль (сплошная/пунктир/точечная), маркеры и толщину.
# Добавьте легенду, подписи осей, сетку и пределы осей так, чтобы все кривые были видны.
x = np.linspace(0,6)
y1 = 2*x + 1
y2 = x**1.5
y3 = 0.5*x**2

plt.plot(x, y1,  color="red", linestyle="--", marker='s', linewidth=1,label="y = 2x + 1")
plt.plot(x, y2, color="blue",linestyle="--", marker='D' , linewidth=2,label="y = x**1.5")
plt.plot(x, y3, color="green",linestyle="-.", marker='*', linewidth=3,label="y = 0.5x**2")

plt.title("Три графика на одной сетке")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.legend()
plt.grid(True)
plt.xlim(0,6)
plt.ylim(0,13)
plt.show()


# Используйте: x = np.linspace(...); plt.plot(..., color=..., linestyle=..., marker=..., linewidth=..., label=...);
# plt.legend(); plt.xlabel(...); plt.ylabel(...); plt.grid(True); plt.xlim(...); plt.ylim(...); plt.show().

```
